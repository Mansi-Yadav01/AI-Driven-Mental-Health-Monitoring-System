```python
# Step 7: Find missing values
missing_values = df.isnull().sum()

# Optionally, display the percentage of missing values
missing_percentage = (df.isnull().sum() / len(df)) * 100

# Display missing values and their percentage
print("Missing values per column:")
print(missing_values)

print("\nPercentage of missing values per column:")
print(missing_percentage)
```

```python
# Step 8: Handle missing values

# Fill missing values in numerical columns with the median
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[numerical_cols] = df[numerical_cols].fillna(df[numerical_cols].median())

# Fill missing values in text (object) columns with 'Unknown'
text_cols = df.select_dtypes(include=['object']).columns
df[text_cols] = df[text_cols].fillna('Unknown')

print("Missing values have been handled.")
```

```python
# Step 9: Drop rows with any remaining missing values
print("Dropping rows with remaining missing values...")
df.dropna(inplace=True)
```

```python
# Step 10: Verify that no missing data remains
print("Verifying that no missing data remains...")
print(df.isnull().sum())
```

```python
# Step 11: Display the structure after handling missing values
print("Dataset structure after handling missing values:")
print(df.info())
```

```python
# Step 12: Convert float columns (except 'BodyTemperature') to integers

# Find all float columns
float_columns = df.select_dtypes(include=['float64']).columns

# Exclude 'BodyTemperature' from conversion
columns_to_convert = [col for col in float_columns if col != 'BodyTemperature']

# Convert these columns to integer type
df[columns_to_convert] = df[columns_to_convert].astype(int)

print("Converted float columns (except 'BodyTemperature') to integers.")
```

```python
#Step 15: Sentiment analysis

import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import pandas as pd

# Download the VADER lexicon
nltk.download('vader_lexicon')

# Initialize the VADER sentiment analyzer
sia = SentimentIntensityAnalyzer()


# Function to calculate sentiment score for each text
def analyze_sentiment(text):
    if pd.isna(text) or len(text.strip()) == 0:  # If the text is NaN or empty, return a neutral score
        return 0
    sentiment_score = sia.polarity_scores(text)  # Get polarity scores from VADER
    return sentiment_score['compound']  # Return the compound score (range -1 to 1)


# Apply sentiment analysis to the text columns
df['PostSentiment'] = df['PostsWithText'].apply(analyze_sentiment)
df['CommentSentiment'] = df['CommentsOnPosts'].apply(analyze_sentiment)
df['StatusSentiment'] = df['StatusUpdates'].apply(analyze_sentiment)

# Display the first few rows of the new sentiment columns
print(df[['PostsWithText', 'PostSentiment', 'CommentsOnPosts', 'CommentSentiment', 'StatusUpdates', 'StatusSentiment']].head())
```

```python
# Function to categorize sentiment into Positive, Neutral, and Negative
def categorize_sentiment(score):
    if score > 0.05:
        return 'Positive'
    elif score < -0.05:
        return 'Negative'
    else:
        return 'Neutral'

# Categorize sentiment into Positive, Negative, Neutral
df['PostSentimentCategory'] = df['PostSentiment'].apply(categorize_sentiment)
df['CommentSentimentCategory'] = df['CommentSentiment'].apply(categorize_sentiment)
df['StatusSentimentCategory'] = df['StatusSentiment'].apply(categorize_sentiment)

# Display the categorized sentiment for inspection
print(df[['PostSentimentCategory', 'CommentSentimentCategory', 'StatusSentimentCategory']].head())
```

```python
#Step 16: Feature Engineering

# 1. Physical Activity Risk Score
df['PhysicalRiskScore'] = ((df['HeartRate'] > 85) | (df['HeartRate'] < 60)).astype(int) + \
                          ((df['SleepHours'] > 9) | (df['SleepHours'] < 3)).astype(int) + \
                          (df['ActivityLevel'] < 700).astype(int) + \
                          (df['BodyTemperature'] > 97).astype(int)

# 2. Routine Disruption Score
df['RoutineDisruptionScore'] = df[['LocationChanges', 'CalendarEvents']].var(axis=1)

# 3. Sentiment-Based Features
# Average Sentiment Score across Posts, Comments, and Status Updates
df['AverageSentiment'] = df[['PostSentiment', 'CommentSentiment', 'StatusSentiment']].mean(axis=1)

# Sentiment Variance
df['SentimentVariance'] = df[['PostSentiment', 'CommentSentiment', 'StatusSentiment']].var(axis=1)

# 4. Social Media Engagement Score
df['EngagementScore'] = df['SocialPosts'] + df['InteractionFrequency'] + df['ScreenTime']

# 5. Mood and Stress-Related Features
# Mood-Stress Ratio
df['MoodStressRatio'] = df['Mood'] / (df['StressLevel'] + 1e-5)  # Small constant to avoid division by zero

# 6. Physical Activity-Related Features
# Active hours estimation
df['ActiveHours'] = df['ActivityLevel'] / 2000  # 2000 steps = 1 hour of activity

# Sleep Deprivation Feature
df['SleepDeprivation'] = np.where(df['SleepHours'] < 3, 1, 0)

# Mobile User (Mobility-based feature)
df['MobileUser'] = np.where(df['LocationChanges'] > 3, 1, 0)

# 7. Interaction of Mood and Sentiment
df['MoodSentimentInteraction'] = df['Mood'] * df['AverageSentiment']


# 8. Social Media Activity Ratio: Ratio of interactions to posts
df['InteractionPostRatio'] = np.where(df['SocialPosts'] > 0, df['InteractionFrequency'] / df['SocialPosts'], 0)

# 9. Stress Prediction Features
# High Stress Indicator: If the stress level is above 7, we classify it as high stress
df['HighStress'] = np.where(df['StressLevel'] > 7, 1, 0)

# Stress Variation: Measure the difference between mood and stress level (higher difference could indicate mental instability)
df['MoodStressDifference'] = df['Mood'] - df['StressLevel']

# 10. Screen Time Related Features
# Binary indicator of excessive screen time (more than 6 hours a day)
df['ExcessiveScreenTime'] = np.where(df['ScreenTime'] > 6, 1, 0)

# Normalized screen time as a percentage of the day (assuming 24 hours in a day)
df['ScreenTimePercentage'] = df['ScreenTime'] / 24


# =======================================
# Verify the new features
# ======================================
print("Newly engineered features:")
print(df[['PhysicalRiskScore', 'RoutineDisruptionScore', 'AverageSentiment', 'EngagementScore',
         'MoodStressRatio', 'ActiveHours', 'SleepDeprivation', 'MobileUser', 'MoodSentimentInteraction',
         'InteractionPostRatio', 'HighStress', 'MoodStressDifference', 'ExcessiveScreenTime',
         'ScreenTimePercentage']].head())
```

```python
# Function to calculate anxiety score based on specific ranges and additional features
def calculate_anxiety_score(row):
    score = 0

    # HeartRate conditions
    if (row['HeartRate'] < 60) or (row['HeartRate'] > 85):  # High Anxiety
        score += 3
    elif (83 <= row['HeartRate'] <= 85):  # Medium Anxiety
        score += 2
    elif (76 <= row['HeartRate'] < 83):  # Low Anxiety
        score += 1
    else:  # No Anxiety
        score += 0

    # SleepHours conditions
    if (row['SleepHours'] < 3) or (row['SleepHours'] > 8):  # High Anxiety
        score += 3
    elif (3 <= row['SleepHours'] <= 4):  # Medium Anxiety
        score += 2
    elif (row['SleepHours'] == 5):  # Low Anxiety
        score += 1
    else:  # No Anxiety
        score += 0

    # ActivityLevel conditions
    if row['ActivityLevel'] < 700:  # High Anxiety
        score += 3
    elif (700 <= row['ActivityLevel'] < 1500):  # Medium Anxiety
        score += 2
    elif (1500 <= row['ActivityLevel'] <= 4000):  # Low Anxiety
        score += 1
    else:  # No Anxiety
        score += 0

    # BodyTemperature conditions
    if row['BodyTemperature'] > 99.0:  # High Anxiety
        score += 3
    elif (99.0 > row['BodyTemperature'] > 97.0):  # Medium Anxiety
        score += 2
    elif (96.0 <= row['BodyTemperature'] <= 97.0):  # Low Anxiety
        score += 1
    else:  # No Anxiety
        score += 0

    # SocialPosts conditions
    if row['SocialPosts'] < 3:  # High Anxiety
        score += 3
    elif (3 <= row['SocialPosts'] < 6):  # Medium Anxiety
        score += 2
    elif (6 <= row['SocialPosts'] <= 10):  # Low Anxiety
        score += 1
    else:  # No Anxiety
        score += 0

    # InteractionFrequency conditions
    if row['InteractionFrequency'] < 25:  # High Anxiety
        score += 3
    elif (25 <= row['InteractionFrequency'] < 40):  # Medium Anxiety
        score += 2
    elif (40 <= row['InteractionFrequency'] <= 50):  # Low Anxiety
        score += 1
    else:  # No Anxiety
        score += 0

    # ScreenTime conditions
    if (row['ScreenTime'] < 2) or (row['ScreenTime'] > 6):  # High Anxiety
        score += 3
    elif (6 <= row['ScreenTime'] <= 7):  # Medium Anxiety
        score += 2
    elif (3 <= row['ScreenTime'] <= 5):  # Low Anxiety
        score += 1
    else:  # No Anxiety
        score += 0

    # LocationChanges conditions
    if (row['LocationChanges'] < 2) or (row['LocationChanges'] > 5):  # High Anxiety
        score += 3
    elif (6 <= row['LocationChanges'] <= 7):  # Medium Anxiety
        score += 2
    elif (4 <= row['LocationChanges'] < 6):  # Low Anxiety
        score += 1
    else:  # No Anxiety
        score += 0

    # CalendarEvents conditions
    if row['CalendarEvents'] > 4:  # High Anxiety
        score += 3
    elif (3 <= row['CalendarEvents'] <= 4):  # Medium Anxiety
        score += 2
    elif (row['CalendarEvents'] ==2):  # Low Anxiety
        score += 1
    else:  # No Anxiety
        score += 0

    # Mood conditions
    if row['Mood'] < 4:  # High Anxiety
        score += 3
    elif (4 <= row['Mood'] < 6):  # Medium Anxiety
        score += 2
    elif (6 <= row['Mood'] < 9):  # Low Anxiety
        score += 1
    else:  # No Anxiety
        score += 0

    # StressLevel conditions
    if row['StressLevel'] > 7:  # High Anxiety
        score += 3
    elif (6 <= row['StressLevel'] <= 7):  # Medium Anxiety
        score += 2
    elif (3 <= row['StressLevel'] < 6):  # Low Anxiety
        score += 1
    else:  # No Anxiety
        score += 0

    # PostSentiment conditions
    if -1.0 <= row['PostSentiment'] <= -0.5:  # High Anxiety
        score += 3
    elif -0.5 < row['PostSentiment'] <= -0.05:  # Medium Anxiety
        score += 2
    elif -0.05 < row['PostSentiment'] <= 0.5:  # Low Anxiety
        score += 1
    else:  # No Anxiety
        score += 0

    # CommentSentiment conditions
    if -1.0 <= row['CommentSentiment'] <= -0.5:  # High Anxiety
        score += 3
    elif -0.5 < row['CommentSentiment'] <= -0.05:  # Medium Anxiety
        score += 2
    elif -0.05 < row['CommentSentiment'] <= 0.5:  # Low Anxiety
        score += 1
    else:  # No Anxiety
        score += 0

    # StatusSentiment conditions
    if -1.0 <= row['StatusSentiment'] <= -0.5:  # High Anxiety
        score += 3
    elif -0.5 < row['StatusSentiment'] <= -0.05:  # Medium Anxiety
        score += 2
    elif -0.05 < row['StatusSentiment'] <= 0.5:  # Low Anxiety
        score += 1
    else:  # No Anxiety
        score += 0

    # PostSentimentCategory conditions
    if row['PostSentimentCategory'] == 'Negative':  # High/Medium Anxiety
        score += 3
    elif row['PostSentimentCategory'] == 'Neutral':  # Low Anxiety
        score += 1
    else:  # No Anxiety
        score += 0

    # CommentSentimentCategory conditions
    if row['CommentSentimentCategory'] == 'Negative':  # High/Medium Anxiety
        score += 3
    elif row['CommentSentimentCategory'] == 'Neutral':  # Low Anxiety
        score += 1
    else:  # No Anxiety
        score += 0

    # StatusSentimentCategory conditions
    if row['StatusSentimentCategory'] == 'Negative':  # High/Medium Anxiety
        score += 3
    elif row['StatusSentimentCategory'] == 'Neutral':  # Low Anxiety
        score += 1
    else:  # No Anxiety
        score += 0

    # Sentiment Variance (High variance can indicate emotional instability)
    if row['SentimentVariance'] > 0.1:  # High variance indicating mood swings
        score += 2
    else:  # No Anxiety for low variance
        score += 0

    # Mood-Stress Ratio (Low ratio indicates imbalance)
    if row['MoodStressRatio'] < 0.5:  # High Anxiety due to imbalance
        score += 3
    elif 0.5 <= row['MoodStressRatio'] < 1.0:  # Medium Anxiety
        score += 2
    else:  # Low Anxiety
        score += 1

    # InteractionPostRatio conditions
    if row['InteractionPostRatio'] < 1:  # High Anxiety (low interactions per post)
        score += 3
    elif 1 <= row['InteractionPostRatio'] <= 2:  # Medium Anxiety
        score += 2
    elif 2 < row['InteractionPostRatio'] <= 3:  # Low Anxiety
        score += 1
    else:  # No Anxiety
        score += 0

    return score

# Apply the function to calculate Anxiety Score for each row
df['AnxietyScore'] = df.apply(calculate_anxiety_score, axis=1)
# Define the function to classify Anxiety Level based on Anxiety Score
def classify_anxiety_level(score):
    if score >= 40:
        return 'High'
    elif 30 <= score < 40:
        return 'Medium'
    elif 15 <= score < 30:
        return 'Low'
    else:
        return 'No'

#  Apply the function to classify Anxiety Level
df['AnxietyLevel'] = df['AnxietyScore'].apply(classify_anxiety_level)
# Function to calculate Depression Score with more focus on mood and social withdrawal, including Anxiety Score
def calculate_depression_score(row):
    score = 0

    # 1. Mood: If Mood is consistently low (below 4), add more to the depression score
    if row['Mood'] < 4:
        score += 4  # Major contributor

    # 2. Activity Level: Low activity can indicate depression (below 1000 steps)
    if row['ActivityLevel'] < 1000:
        score += 2  # Moderate contributor

    # 3. SleepHours: Too little (below 4) or too much sleep (above 9) can indicate depression
    if (row['SleepHours'] < 4) or (row['SleepHours'] > 9):
        score += 2  # Moderate contributor

    # 4. Social Posts: Fewer posts indicate potential social withdrawal
    if row['SocialPosts'] < 3:
        score += 2  # Moderate contributor

    # 5. Interaction Frequency: Fewer interactions indicate social withdrawal
    if row['InteractionFrequency'] < 20:
        score += 2  # Moderate contributor

    # Add the Anxiety Score to the Depression Score
    total_depression_score = score + row['AnxietyScore']

    return total_depression_score

# Apply the function to calculate Depression Score for each row
df['DepressionScore'] = df.apply(calculate_depression_score, axis=1)
#  Define the function to classify Depression Level based on Depression Score
def classify_depression_level(score):
    if score >= 50:
        return 'High'
    elif 40 <= score < 50:
        return 'Medium'
    elif 30 <= score < 40:
        return 'Low'
    else:
        return 'No'

#  Apply the function to classify Depression Level
df['DepressionLevel'] = df['DepressionScore'].apply(classify_depression_level)
# Display the first few rows to verify
print(df[['HeartRate', 'SleepHours', 'ActivityLevel', 'PostSentiment', 'CommentSentiment',
          'StatusSentiment', 'AverageSentiment', 'SentimentVariance', 'AnxietyScore', 'AnxietyLevel', 'DepressionScore', 'DepressionLevel', 'HighStress']].head())
```

```python
#Step 19: model development

#data preparation for model development
from sklearn.model_selection import train_test_split

# Features for model development (select relevant columns including sentiment columns)
features = ['HeartRate', 'SleepHours', 'ActivityLevel', 'BodyTemperature', 'SocialPosts',
            'InteractionFrequency', 'ScreenTime', 'LocationChanges', 'CalendarEvents',
            'Mood', 'StressLevel','PostSentiment','CommentSentiment', 'StatusSentiment',
            'RoutineDisruptionScore',
            'AverageSentiment', 'SentimentVariance', 'EngagementScore', 'MoodStressRatio',
            'ActiveHours', 'SleepDeprivation', 'MobileUser', 'MoodSentimentInteraction',
            'InteractionPostRatio', 'HighStress', 'MoodStressDifference', 'ExcessiveScreenTime',
            'ScreenTimePercentage']

# Split data for Anxiety Score (regression)
X_anxiety_score = df[features]
y_anxiety_score = df['AnxietyScore']

# Split data for Anxiety Level (classification)
X_anxiety_level = df[features]
y_anxiety_level = df['AnxietyLevel']

# Split data for Depression Score (regression)
X_depression_score = df[features]
y_depression_score = df['DepressionScore']

# Split data for Depression Level (classification)
X_depression_level = df[features]
y_depression_level = df['DepressionLevel']

# Split data for High Stress (classification)
X_high_stress = df[features]
y_high_stress = df['HighStress']

# Train-test split for each task (80% train, 20% test)
X_train_anxiety_score, X_test_anxiety_score, y_train_anxiety_score, y_test_anxiety_score = train_test_split(X_anxiety_score, y_anxiety_score, test_size=0.2, random_state=42)
X_train_anxiety_level, X_test_anxiety_level, y_train_anxiety_level, y_test_anxiety_level = train_test_split(X_anxiety_level, y_anxiety_level, test_size=0.2, random_state=42)
X_train_depression_score, X_test_depression_score, y_train_depression_score, y_test_depression_score = train_test_split(X_depression_score, y_depression_score, test_size=0.2, random_state=42)
X_train_depression_level, X_test_depression_level, y_train_depression_level, y_test_depression_level = train_test_split(X_depression_level, y_depression_level, test_size=0.2, random_state=42)
X_train_high_stress, X_test_high_stress, y_train_high_stress, y_test_high_stress = train_test_split(X_high_stress, y_high_stress, test_size=0.2, random_state=42)

print("Data preparation complete.")
```

```python
#model selection and definition

from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LinearRegression, LogisticRegression

# Classification models
anxiety_level_classifier = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
depression_level_classifier = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
high_stress_classifier = LogisticRegression(max_iter=100, random_state=42)

# Regression models
anxiety_score_regressor = LinearRegression()
depression_score_regressor = LinearRegression()

print("Models are selected and defined for classification and regression tasks.")
```

```python
# Train the models

# 1. Train regression models (Anxiety Score, Depression Score)
anxiety_score_regressor.fit(X_train_anxiety_score, y_train_anxiety_score)
depression_score_regressor.fit(X_train_depression_score, y_train_depression_score)

# 2. Train classification models (Anxiety Level, Depression Level, High Stress)
anxiety_level_classifier.fit(X_train_anxiety_level, y_train_anxiety_level)
depression_level_classifier.fit(X_train_depression_level, y_train_depression_level)
high_stress_classifier.fit(X_train_high_stress, y_train_high_stress)

print("Models are trained successfully.")
```

```python
#model evaluation

from sklearn.metrics import accuracy_score, classification_report, mean_squared_error, confusion_matrix

# 1. Evaluate Regression Models (Anxiety Score, Depression Score)
# Predictions for Anxiety Score
y_pred_anxiety_score = anxiety_score_regressor.predict(X_test_anxiety_score)
anxiety_score_rmse = mean_squared_error(y_test_anxiety_score, y_pred_anxiety_score, squared=False)

# Predictions for Depression Score
y_pred_depression_score = depression_score_regressor.predict(X_test_depression_score)
depression_score_rmse = mean_squared_error(y_test_depression_score, y_pred_depression_score, squared=False)

print(f"Anxiety Score RMSE: {anxiety_score_rmse}")
print(f"Depression Score RMSE: {depression_score_rmse}")

# 2. Evaluate Classification Models (Anxiety Level, Depression Level, High Stress)
# Predictions for Anxiety Level
y_pred_anxiety_level = anxiety_level_classifier.predict(X_test_anxiety_level)
anxiety_level_accuracy = accuracy_score(y_test_anxiety_level, y_pred_anxiety_level)
print(f"Anxiety Level Accuracy: {anxiety_level_accuracy}")
print(classification_report(y_test_anxiety_level, y_pred_anxiety_level))

# Predictions for Depression Level
y_pred_depression_level = depression_level_classifier.predict(X_test_depression_level)
depression_level_accuracy = accuracy_score(y_test_depression_level, y_pred_depression_level)
print(f"Depression Level Accuracy: {depression_level_accuracy}")
print(classification_report(y_test_depression_level, y_pred_depression_level))

# Predictions for High Stress
y_pred_high_stress = high_stress_classifier.predict(X_test_high_stress)
high_stress_accuracy = accuracy_score(y_test_high_stress, y_pred_high_stress)
print(f"High Stress Accuracy: {high_stress_accuracy}")
print(classification_report(y_test_high_stress, y_pred_high_stress))
```

```python
#Hyperparameter tuning

from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import accuracy_score, mean_squared_error
import numpy as np

# Define the models
anxiety_level_classifier = RandomForestClassifier(random_state=42)
depression_level_classifier = RandomForestClassifier(random_state=42)
anxiety_score_regressor = LinearRegression()
depression_score_regressor = LinearRegression()
high_stress_classifier = LogisticRegression()

# Hyperparameter tuning for RandomForestClassifier (Anxiety Level)
param_dist_class = {
    'n_estimators': [50, 100],  # Fewer estimators for faster training
    'max_depth': [5, 10],  # Reduced depth to speed up training
    'min_samples_split': [2, 5]
}

# RandomizedSearchCV for Anxiety Level Classifier
random_search_anxiety_level = RandomizedSearchCV(
    estimator=anxiety_level_classifier, param_distributions=param_dist_class,
    n_iter=5, cv=2, scoring='accuracy', random_state=42, n_jobs=-1
)
random_search_anxiety_level.fit(X_train_anxiety_level, y_train_anxiety_level)

# RandomizedSearchCV for Depression Level Classifier
random_search_depression_level = RandomizedSearchCV(
    estimator=depression_level_classifier, param_distributions=param_dist_class,
    n_iter=5, cv=2, scoring='accuracy', random_state=42, n_jobs=-1
)
random_search_depression_level.fit(X_train_depression_level, y_train_depression_level)

# Hyperparameter tuning for LinearRegression (Anxiety Score)
param_dist_reg = {
    'fit_intercept': [True, False]
}

# RandomizedSearchCV for Anxiety Score Regressor
random_search_anxiety_score = RandomizedSearchCV(
    estimator=anxiety_score_regressor, param_distributions=param_dist_reg,
    n_iter=4, cv=2, scoring='neg_mean_squared_error', random_state=42, n_jobs=-1
)
random_search_anxiety_score.fit(X_train_anxiety_score, y_train_anxiety_score)

# RandomizedSearchCV for Depression Score Regressor
random_search_depression_score = RandomizedSearchCV(
    estimator=depression_score_regressor, param_distributions=param_dist_reg,
    n_iter=4, cv=2, scoring='neg_mean_squared_error', random_state=42, n_jobs=-1
)
random_search_depression_score.fit(X_train_depression_score, y_train_depression_score)

# Hyperparameter tuning for LogisticRegression (High Stress Classification)
param_dist_logreg = {
    'C': [0.1, 1.0, 10],  # Regularization strength
    'max_iter': [100, 200]  # Number of iterations
}

# RandomizedSearchCV for High Stress Classifier
random_search_high_stress = RandomizedSearchCV(
    estimator=high_stress_classifier, param_distributions=param_dist_logreg,
    n_iter=5, cv=2, scoring='accuracy', random_state=42, n_jobs=-1
)
random_search_high_stress.fit(X_train_high_stress, y_train_high_stress)

# Get the best hyperparameters for each model
print("Best parameters for Anxiety Level Classifier:", random_search_anxiety_level.best_params_)
print("Best parameters for Depression Level Classifier:", random_search_depression_level.best_params_)
print("Best parameters for Anxiety Score Regressor:", random_search_anxiety_score.best_params_)
print("Best parameters for Depression Score Regressor:", random_search_depression_score.best_params_)
print("Best parameters for High Stress Classifier:", random_search_high_stress.best_params_)

# Evaluate the models on test data
# Anxiety Level Classifier
y_pred_anxiety_level = random_search_anxiety_level.predict(X_test_anxiety_level)
anxiety_level_accuracy = accuracy_score(y_test_anxiety_level, y_pred_anxiety_level)
print("Anxiety Level Classifier Accuracy:", anxiety_level_accuracy)

# Depression Level Classifier
y_pred_depression_level = random_search_depression_level.predict(X_test_depression_level)
depression_level_accuracy = accuracy_score(y_test_depression_level, y_pred_depression_level)
print("Depression Level Classifier Accuracy:", depression_level_accuracy)

# Anxiety Score Regressor
y_pred_anxiety_score = random_search_anxiety_score.predict(X_test_anxiety_score)
anxiety_score_rmse = np.sqrt(mean_squared_error(y_test_anxiety_score, y_pred_anxiety_score))
print("Anxiety Score Regressor RMSE:", anxiety_score_rmse)

# Depression Score Regressor
y_pred_depression_score = random_search_depression_score.predict(X_test_depression_score)
depression_score_rmse = np.sqrt(mean_squared_error(y_test_depression_score, y_pred_depression_score))
print("Depression Score Regressor RMSE:", depression_score_rmse)

# High Stress Classifier
y_pred_high_stress = random_search_high_stress.predict(X_test_high_stress)
high_stress_accuracy = accuracy_score(y_test_high_stress, y_pred_high_stress)
print("High Stress Classifier Accuracy:", high_stress_accuracy)
```
